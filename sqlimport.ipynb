```python
import pandas as pd
import mysql.connector

# Load CSV file
file_path = "Tesla-Lithium Dataset_Tesla_Lithium_Dataset.csv"
df = pd.read_csv(file_path)

# ✅ Convert Date column to MySQL-friendly format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",        # change if needed
    user="root",             # replace with your MySQL username
    password="sql1234",      # replace with your MySQL password
    database="tesla_lithium_db"
)
cursor = conn.cursor()

# Insert rows into table
sql = """
INSERT INTO tesla_lithium_data 
(date, `Tesla Sales per day (100)`, 
 `Lithium Mining Supply Pounds/Day (50 lbs)`, 
 `Tesla Car Price (500 k $)`, 
 `Lithium Price/Pound`, 
 `Lithium_Comp_share`, 
 `Tesla_Share`)
VALUES (%s, %s, %s, %s, %s, %s, %s)
"""

# ✅ Use executemany for speed
values_list = [
    (
        row['Date'],
        int(row['Tesla Sales per day (100)']),
        int(row['Lithium Mining Supply Pounds/Day (50 lbs)']),
        float(row['Tesla Car Price (500 k $)']),
        float(row['Lithium Price/Pound']),
        float(row['Lithium_Comp_share']),
        float(row['Tesla_Share'])
    )
    for _, row in df.iterrows()
]

cursor.executemany(sql, values_list)

# Commit changes and close connection
conn.commit()
cursor.close()
conn.close()

print("Data successfully inserted into tesla_lithium_data table.")
```
